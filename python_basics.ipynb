Data types means what kind of value store in variables. Python has various data types including integers, floats, strings, and booleans.

# 3. Operators
An operator in Python is a special symbol or keyword used to perform operations on values or variables.
Python uses operators to carry out computations like arithmatic, comparison, logical evaluation, assignment, Relational, Bitwise, Membership and Identity etc.

| **Category**                | **Operator**    | **Description**               | **Example**                         |
| --------------------------- | --------------- | ----------------------------- | ----------------------------------- |
| **Arithmetic**              | `+`             | Addition                      | `a + b`                             |
| Arithmetic                  | `-`             | Subtraction                   | `a - b`                             |
| Arithmetic                  | `*`             | Multiplication                | `a * b`                             |
| Arithmetic                  | `/`             | Division                      | `a / b`                             |
| Arithmetic                  | `%`             | Modulus (remainder)           | `a % b`                             |
| Arithmetic                  | `**`            | Exponentiation                | `a ** b`                            |
| Arithmetic                  | `//`            | Floor Division                | `a // b`                            |
| **Relational (Comparison)** | `==`            | Equal to                      | `a == b`                            |
| Comparison                  | `!=`            | Not equal to                  | `a != b`                            |
| Comparison                  | `>`             | Greater than                  | `a > b`                             |
| Comparison                  | `<`             | Less than                     | `a < b`                             |
| Comparison                  | `>=`            | Greater than or equal         | `a >= b`                            |
| Comparison                  | `<=`            | Less than or equal            | `a <= b`                            |
| **Logical**                 | `and`           | True if both True             | `a > 5 and b < 10`                  |
| Logical                     | `or`            | True if any True              | `a > 5 or b < 10`                   |
| Logical                     | `not`           | Negates condition             | `not(a > b)`                        |
| **Assignment**              | `=`             | Assign value                  | `a = 10`                            |
| Assignment                  | `+=`            | Add and assign                | `a += 2`                            |
| Assignment                  | `-=`            | Subtract and assign           | `a -= 2`                            |
| Assignment                  | `*=`            | Multiply and assign           | `a *= 2`                            |
| Assignment                  | `/=`            | Divide and assign             | `a /= 2`                            |
| Assignment                  | `%=`            | Modulus and assign            | `a %= 2`                            |
| Assignment                  | `**=`           | Exponent and assign           | `a **= 2`                           |
| Assignment                  | `//=`           | Floor divide and assign       | `a //= 2`                           |
| **Bitwise**                 | `&`             | Bitwise AND                   | `a & b`                             |
| Bitwise                     | `\|`            | Bitwise OR                    | `a \| b`                            |
| Bitwise                     | `^`             | Bitwise XOR                   | `a ^ b`                             |
| Bitwise                     | `~`             | Bitwise NOT                   | `~a`                                |
| Bitwise                     | `<<`            | Left shift                    | `a << 2`                            |
| Bitwise                     | `>>`            | Right shift                   | `a >> 2`                            |
| **Membership**              | `in`            | True if value in sequence     | `'a' in 'apple'`                    |
| Membership                  | `not in`        | True if value not in sequence | `'x' not in 'apple'`                |
| **Identity**                | `is`            | Same object                   | `a is b`                            |
| Identity                    | `is not`        | Not same object               | `a is not b`                        |
| **Ternary (Conditional)**   | `a if c else b` | One-line if-else              | `"Even" if num % 2 == 0 else "Odd"` |

# 4. Conditional Statement

| Statement / Syntax                      | Short Description                                        |
| --------------------------------------- | -------------------------------------------------------- |
| `if`                                    | Executes a block of code when the condition is **True**. |
| `if-else`                               | Chooses between two blocks depending on **True/False**.  |
| `if-elif-else`                          | Checks multiple conditions in sequence.                  |
| Nested `if`                             | An `if` statement inside another `if` block.             |
| Ternary expression (`a if cond else b`) | Short, one-line conditional expression.                  |


```python
# if
age=15
if age>=18:
    print("You are eligible for Voting")
```

```python
#if-else
if age>=18:
    print("You are eligible for Voting")
else:
    print("You are not eligible for Voting")
```

```python
#if-elif-else
marks=80
if marks < 40:
    print("you are Failed")
elif 40 <= marks <60:
    print("you are pass with second class")
elif 60<= marks <75:
    print("you are pass with first class")
elif 75<= marks <100:
    print("You are pass with first class distinction")
else:
    print("value was incorrect")
```

```python
#Nested if statement
state="Maharshtra"
dist="Latur"
if state == "Maharshtra":
    if dist== "Chh. Sambhajinagar":
        print("You are from Chh. Sambhajinagar in Maharashtra")
    if dist== "Latur":
        print("You are from Latur in Maharashtra")
```

```python
#Ternary Statement
n=17
"Even" if n%2==0 else "odd"
```

# 5. Loops and Control Statements

| Loop / Statement | Short Description                                 |
| ---------------- | ------------------------------------------------- |
| `for` loop       | Loops over a sequence (list, range, string, etc.) |
| `while` loop     | Repeats as long as the condition is `True`        |
| `break`          | Exits the loop immediately                        |
| `continue`       | Skips the current iteration                       |
| `pass`           | Does nothing; placeholder                         |


```python
#For Loops
s1="Hello"
list1=[1,2,3,"gopal",5]
#string iteration using for Loop
for c in s1:
    print("iterate value:",c)
print("--------------partition-----------------")
for item in list1:
    print("iterated item:",item)
```

```python
#Numberic Value
for n in range(10):
    print("interated number: ",n)
```

```python
#While Loop
n=10
while 0<n:
    print("Condition satisfied",n)
    n=n-1
```

```python
for i in range(10):
    try:
        a=1/1
        print("Retry:",i)
        break
    except:
        pass
```

```python
#Control Statement
#Break
for i in range(10):
    if i == 5:
        break
    print("range number: ",i)
```

```python
for i in range(10):
    if i%2!=0:
        continue
    print("range number: ",i)
```
