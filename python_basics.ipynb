# 3. Operators
An operator in Python is a special symbol or keyword used to perform operations on values or variables.
Python uses operators to carry out computations like arithmatic, comparison, logical evaluation, assignment, Relational, Bitwise, Membership and Identity etc.

| **Category**                | **Operator**    | **Description**               | **Example**                         |
| --------------------------- | --------------- | ----------------------------- | ----------------------------------- |

| Arithmetic                  | `%`             | Modulus (remainder)           | `a % b`                             |

# 4. Conditional Statement

| Statement / Syntax                      | Short Description                                        |
| --------------------------------------- | -------------------------------------------------------- |
| `if`                                    | Executes a block of code when the condition is **True**. |
| `if-else`                               | Chooses between two blocks depending on **True/False**.  |
| `if-elif-else`                          | Checks multiple conditions in sequence.                  |
| Nested `if`                             | An `if` statement inside another `if` block.             |
| Ternary expression (`a if cond else b`) | Short, one-line conditional expression.                  |


```python
# if
age=15
if age>=18:
    print("You are eligible for Voting")
```

```python
#if-else
if age>=18:
    print("You are eligible for Voting")
else:
    print("You are not eligible for Voting")
```

```python
#if-elif-else
marks=80
if marks < 40:
    print("you are Failed")
elif 40 <= marks <60:
    print("you are pass with second class")
elif 60<= marks <75:
    print("you are pass with first class")
elif 75<= marks <100:
    print("You are pass with first class distinction")
else:
    print("value was incorrect")
```

```python
#Nested if statement
state="Maharshtra"
dist="Latur"
if state == "Maharshtra":
    if dist== "Chh. Sambhajinagar":
        print("You are from Chh. Sambhajinagar in Maharashtra")
    if dist== "Latur":
        print("You are from Latur in Maharashtra")
```

```python
#Ternary Statement
n=17
"Even" if n%2==0 else "odd"
```

## 6.3 Sets
A set in Python is an unordered, mutable collection of unique elements.
It does not allow duplicates and does not maintain order.

```python
set1={1,3,5,4,2,4,4,4}
print(set1)
```

```python
type(set1)
```

| Method                              | Description                               | Application                      | Example (Input → Output)                                  |
| ----------------------------------- | ----------------------------------------- | -------------------------------- | --------------------------------------------------------- |
| **add(x)**                          | Adds element `x` to the set               | Insert new item                  | `{1,2}.add(3)` → `{1,2,3}`                                |
| **clear()**                         | Removes all elements                      | Reset set                        | `{1,2}.clear()` → `set()`                                 |
| **copy()**                          | Returns shallow copy                      | Duplicate set                    | `s.copy()` → `{…}`                                        |
| **difference(s2)**                  | Elements in s1 but not in s2              | Find non-overlapping items       | `{1,2,3}.difference({2,3})` → `{1}`                       |
| **difference_update(s2)**           | Removes elements of s2 from original set  | In-place difference              | `s={1,2,3}; s.difference_update({2}); s` → `{1,3}`        |
| **discard(x)**                      | Removes `x` if exists (no error)          | Safe removal                     | `{1,2}.discard(2)` → `{1}`                                |
| **intersection(s2)**                | Elements common in both sets              | Find shared items                | `{1,2,3}.intersection({2,3,4})` → `{2,3}`                 |
| **intersection_update(s2)**         | Keeps only common elements (in-place)     | In-place intersection            | `s={1,2,3}; s.intersection_update({2,3})` → `{2,3}`       |
| **isdisjoint(s2)**                  | True if no common elements                | Check if two sets overlap        | `{1,2}.isdisjoint({3})` → `True`                          |
| **issubset(s2)**                    | Checks if set is subset of s2             | Check containment                | `{1,2}.issubset({1,2,3})` → `True`                        |
| **issuperset(s2)**                  | Checks if set is superset of s2           | Check if s contains all elements | `{1,2,3}.issuperset({1})` → `True`                        |
| **pop()**                           | Removes and returns random element        | Remove unknown element           | `{1,2,3}.pop()` → e.g. `1`                                |
| **remove(x)**                       | Removes `x`; raises error if absent       | Strict removal                   | `{1,2}.remove(2)` → `{1}`                                 |
| **symmetric_difference(s2)**        | Returns elements in either set, not both  | XOR set operation                | `{1,2}.symmetric_difference({2,3})` → `{1,3}`             |
| **symmetric_difference_update(s2)** | Updates set with symmetric difference     | In-place XOR                     | `s={1,2}; s.symmetric_difference_update({2,3})` → `{1,3}` |
| **union(s2,... )**                  | Returns all unique elements from all sets | Combine many sets                | `{1,2}.union({2,3},{3,4})` → `{1,2,3,4}`                  |
| **update(s2)**                      | Adds all elements of s2                   | In-place merge                   | `s={1,2}; s.update({2,3})` → `{1,2,3}`                    |

| Method                        | Description                                                    | Typical Application                       | Example (Input → Output)                                         |
| ----------------------------- | -------------------------------------------------------------- | ----------------------------------------- | ---------------------------------------------------------------- |
| **clear()**                   | Removes all items from the dictionary                          | Reset a dictionary to empty               | `d = {'a':1,'b':2}; d.clear(); print(d)` → `{}`                  |
| **copy()**                    | Returns a shallow copy of the dictionary                       | Duplicate dict without linking references | `d={'x':1}; c=d.copy(); print(c)` → `{'x':1}`                    |
| **fromkeys(seq, value=None)** | Creates a new dict with keys from sequence and one value       | Initialize dictionary with default values | `d=dict.fromkeys(['a','b'],0); print(d)` → `{'a':0,'b':0}`       |
| **get(key, default=None)**    | Returns value for key; returns default if key not found        | Safe key access without error             | `d={'a':1}; print(d.get('b',0))` → `0`                           |
| **items()**                   | Returns a view object of `(key, value)` pairs                  | Iterating through items                   | `d={'a':1}; print(list(d.items()))` → `[('a',1)]`                |
| **keys()**                    | Returns a view of dictionary keys                              | Loop through keys                         | `d={'a':1}; print(list(d.keys()))` → `['a']`                     |
| **pop(key, default)**         | Removes key and returns its value; returns default if missing  | Remove and get item at same time          | `d={'a':1}; print(d.pop('a'))` → `1`                             |
| **popitem()**                 | Removes and returns last inserted `(key, value)` pair          | LIFO removal (stack-like)                 | `d={'a':1,'b':2}; print(d.popitem())` → `('b',2)`                |
| **setdefault(key, default)**  | Returns value of key; inserts key with default if missing      | Insert if absent, otherwise get           | `d={}; print(d.setdefault('a',10)); print(d)` → `10`, `{'a':10}` |
| **update(other_dict)**        | Updates the dictionary with other dictionary or iterable pairs | Merge dictionaries                        | `d={'a':1}; d.update({'b':2}); print(d)` → `{'a':1,'b':2}`       |
| **values()**                  | Returns a view of dictionary values                            | Loop through values                       | `d={'a':1}; print(list(d.values()))` → `[1]`                     |

```python
def check_results(marks):
    if marks < 40:
        print("you are Failed")
    elif 40 <= marks <60:
        print("you are pass with second class")
    elif 60<= marks <75:
        print("you are pass with first class")
    elif 75<= marks <100:
        print("You are pass with first class distinction")
    else:
        print("value was incorrect")
student1_marks=72
student2_marks=81
student3_marks=34
check_results(student1_marks)
check_results(student2_marks)
check_results(student3_marks)
```

### *args : 
This Allow a function to accept any number of input positional arguement

```python
def add_number(*args):
    return sum(args)
add_number(2,3,4,5,6,7,1,8,5,3,4)
```

### **kwargs:
This Allow function to accept any keyword arguemnt.

```python
def print_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")
    
print_info(name="Jon",age=30,city="newyork")
```
