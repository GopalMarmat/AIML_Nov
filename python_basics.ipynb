## 1. Variables and Data Types
A variable is a named place in memory where a value is stored so you can use or change it later, Variables are containers for storing data values. 

Data types means what kind of value store in variables. Python has various data types including integers, floats, strings, and booleans.

```python
#Code
n= 23
a="gopal"
print(type(n),type(a))
```

```python
# DataType
a=25 #int
s= "Gopal" #string
f=99.99 #Float
b=True #boolean
print(type(a))
print(type(s))
print(type(f))
print(type(b))
```

## 2. Type Casting
Convert from one data type to another using built-in functions like `int()`, `float()`, `str()`.

```python
#Type Conversion
n=25
print(type(n))
n=str(n) #Type Conversion from integer to string
print(type(n))
```

```python
s="25"
print(type(s))
s=int(s) #Type Conversion from string to integer
print(type(s))
```

```python
n=25
print(type(n))
print(n)
n=float(n) # Type Conversion from Interger to Float
print(type(n))
print(n)
```

# 3. Operators
An operator in Python is a special symbol or keyword used to perform operations on values or variables.
Python uses operators to carry out computations like arithmatic, comparison, logical evaluation, assignment, Relational, Bitwise, Membership and Identity etc.

| **Category**                | **Operator**    | **Description**               | **Example**                         |
| --------------------------- | --------------- | ----------------------------- | ----------------------------------- |
| **Arithmetic**              | `+`             | Addition                      | `a + b`                             |
| Arithmetic                  | `-`             | Subtraction                   | `a - b`                             |
| Arithmetic                  | `*`             | Multiplication                | `a * b`                             |
| Arithmetic                  | `/`             | Division                      | `a / b`                             |
| Arithmetic                  | `%`             | Modulus (remainder)           | `a % b`                             |
| Arithmetic                  | `**`            | Exponentiation                | `a ** b`                            |
| Arithmetic                  | `//`            | Floor Division                | `a // b`                            |
| **Relational (Comparison)** | `==`            | Equal to                      | `a == b`                            |
| Comparison                  | `!=`            | Not equal to                  | `a != b`                            |
| Comparison                  | `>`             | Greater than                  | `a > b`                             |
| Comparison                  | `<`             | Less than                     | `a < b`                             |
| Comparison                  | `>=`            | Greater than or equal         | `a >= b`                            |
| Comparison                  | `<=`            | Less than or equal            | `a <= b`                            |
| **Logical**                 | `and`           | True if both True             | `a > 5 and b < 10`                  |
| Logical                     | `or`            | True if any True              | `a > 5 or b < 10`                   |
| Logical                     | `not`           | Negates condition             | `not(a > b)`                        |
| **Assignment**              | `=`             | Assign value                  | `a = 10`                            |
| Assignment                  | `+=`            | Add and assign                | `a += 2`                            |
| Assignment                  | `-=`            | Subtract and assign           | `a -= 2`                            |
| Assignment                  | `*=`            | Multiply and assign           | `a *= 2`                            |
| Assignment                  | `/=`            | Divide and assign             | `a /= 2`                            |
| Assignment                  | `%=`            | Modulus and assign            | `a %= 2`                            |
| Assignment                  | `**=`           | Exponent and assign           | `a **= 2`                           |
| Assignment                  | `//=`           | Floor divide and assign       | `a //= 2`                           |
| **Bitwise**                 | `&`             | Bitwise AND                   | `a & b`                             |
| Bitwise                     | `\|`            | Bitwise OR                    | `a \| b`                            |
| Bitwise                     | `^`             | Bitwise XOR                   | `a ^ b`                             |
| Bitwise                     | `~`             | Bitwise NOT                   | `~a`                                |
| Bitwise                     | `<<`            | Left shift                    | `a << 2`                            |
| Bitwise                     | `>>`            | Right shift                   | `a >> 2`                            |
| **Membership**              | `in`            | True if value in sequence     | `'a' in 'apple'`                    |
| Membership                  | `not in`        | True if value not in sequence | `'x' not in 'apple'`                |
| **Identity**                | `is`            | Same object                   | `a is b`                            |
| Identity                    | `is not`        | Not same object               | `a is not b`                        |
| **Ternary (Conditional)**   | `a if c else b` | One-line if-else              | `"Even" if num % 2 == 0 else "Odd"` |
