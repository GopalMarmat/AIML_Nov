# 3. Operators
An operator in Python is a special symbol or keyword used to perform operations on values or variables.
Python uses operators to carry out computations like arithmatic, comparison, logical evaluation, assignment, Relational, Bitwise, Membership and Identity etc.

| **Category**                | **Operator**    | **Description**               | **Example**                         |
| --------------------------- | --------------- | ----------------------------- | ----------------------------------- |

| Arithmetic                  | `%`             | Modulus (remainder)           | `a % b`                             |

# 4. Conditional Statement

| Statement / Syntax                      | Short Description                                        |
| --------------------------------------- | -------------------------------------------------------- |
| `if`                                    | Executes a block of code when the condition is **True**. |
| `if-else`                               | Chooses between two blocks depending on **True/False**.  |
| `if-elif-else`                          | Checks multiple conditions in sequence.                  |
| Nested `if`                             | An `if` statement inside another `if` block.             |
| Ternary expression (`a if cond else b`) | Short, one-line conditional expression.                  |


```python
# if
age=15
if age>=18:
    print("You are eligible for Voting")
```

```python
#if-else
if age>=18:
    print("You are eligible for Voting")
else:
    print("You are not eligible for Voting")
```

```python
#if-elif-else
marks=80
if marks < 40:
    print("you are Failed")
elif 40 <= marks <60:
    print("you are pass with second class")
elif 60<= marks <75:
    print("you are pass with first class")
elif 75<= marks <100:
    print("You are pass with first class distinction")
else:
    print("value was incorrect")
```

```python
#Nested if statement
state="Maharshtra"
dist="Latur"
if state == "Maharshtra":
    if dist== "Chh. Sambhajinagar":
        print("You are from Chh. Sambhajinagar in Maharashtra")
    if dist== "Latur":
        print("You are from Latur in Maharashtra")
```

```python
#Ternary Statement
n=17
"Even" if n%2==0 else "odd"
```

## 6.3 Sets
A set in Python is an unordered, mutable collection of unique elements.
It does not allow duplicates and does not maintain order.

```python
set1={1,3,5,4,2,4,4,4}
print(set1)
```

```python
type(set1)
```

| Method                              | Description                               | Application                      | Example (Input → Output)                                  |
| ----------------------------------- | ----------------------------------------- | -------------------------------- | --------------------------------------------------------- |
| **add(x)**                          | Adds element `x` to the set               | Insert new item                  | `{1,2}.add(3)` → `{1,2,3}`                                |
| **clear()**                         | Removes all elements                      | Reset set                        | `{1,2}.clear()` → `set()`                                 |
| **copy()**                          | Returns shallow copy                      | Duplicate set                    | `s.copy()` → `{…}`                                        |
| **difference(s2)**                  | Elements in s1 but not in s2              | Find non-overlapping items       | `{1,2,3}.difference({2,3})` → `{1}`                       |
| **difference_update(s2)**           | Removes elements of s2 from original set  | In-place difference              | `s={1,2,3}; s.difference_update({2}); s` → `{1,3}`        |
| **discard(x)**                      | Removes `x` if exists (no error)          | Safe removal                     | `{1,2}.discard(2)` → `{1}`                                |
| **intersection(s2)**                | Elements common in both sets              | Find shared items                | `{1,2,3}.intersection({2,3,4})` → `{2,3}`                 |
| **intersection_update(s2)**         | Keeps only common elements (in-place)     | In-place intersection            | `s={1,2,3}; s.intersection_update({2,3})` → `{2,3}`       |
| **isdisjoint(s2)**                  | True if no common elements                | Check if two sets overlap        | `{1,2}.isdisjoint({3})` → `True`                          |
| **issubset(s2)**                    | Checks if set is subset of s2             | Check containment                | `{1,2}.issubset({1,2,3})` → `True`                        |
| **issuperset(s2)**                  | Checks if set is superset of s2           | Check if s contains all elements | `{1,2,3}.issuperset({1})` → `True`                        |
| **pop()**                           | Removes and returns random element        | Remove unknown element           | `{1,2,3}.pop()` → e.g. `1`                                |
| **remove(x)**                       | Removes `x`; raises error if absent       | Strict removal                   | `{1,2}.remove(2)` → `{1}`                                 |
| **symmetric_difference(s2)**        | Returns elements in either set, not both  | XOR set operation                | `{1,2}.symmetric_difference({2,3})` → `{1,3}`             |
| **symmetric_difference_update(s2)** | Updates set with symmetric difference     | In-place XOR                     | `s={1,2}; s.symmetric_difference_update({2,3})` → `{1,3}` |
| **union(s2,... )**                  | Returns all unique elements from all sets | Combine many sets                | `{1,2}.union({2,3},{3,4})` → `{1,2,3,4}`                  |
| **update(s2)**                      | Adds all elements of s2                   | In-place merge                   | `s={1,2}; s.update({2,3})` → `{1,2,3}`                    |

## 7.3 Generator Function
Generator function is same like normal function but in generator function we generate the value using yield keyword.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
        gen = count_up_to(5)
        for i in gen:
            print(i)
count_up_to(2)
```

## 7.4 Iterator
An iterator is an object that implements the __iter__() and __next__() methods, allowing traversal through a collection (like a list, tuple, or custom objects).

Key Points:

An iterator must implement __iter__() (which returns the iterator object itself) and __next__() (which returns the next item).

Iterators are used in loops like for.

```python
class count:
    def __init__(self,start):
        self.start =start
    
    def __iter__(self):
        self.count=self.start
        return self
    def __next__(self):
        if self.count > 0:
            self.count -=1
            return self.count
        else:
            raise StopIteration
        
count=count(4)
for num in count:
    print(num)
```

## 7.5 Recursion
Recursion is a process in which a function calls itself to solve smaller instances of the same problem. Base cases are essential to prevent infinite recursion.

Key Points:

Recursion involves a function calling itself.

It requires a base case to stop the recursion.

It is useful for problems that can be divided into smaller subproblems.

```python
# Factorial Code
def recursion_fun(n):
    print(n)
    if n==1:
        return 1
    else:
        return n*recursion_fun(n-1)
recursion_fun(6)
```

# 8 Try-Except
try and except are used for exception handling in Python. The code inside the try block is executed, and if an error occurs, the code in the except block is run.

Key Points:

The try block contains code that may raise an exception.

The except block handles specific exceptions.

Optionally, else can be used to run code if no exception occurs, and finally ensures certain code runs regardless of exceptions.

```python
def try_except(n1,num):
    try:
        n=n1/num
        return n
    except:
        print("number is not divisible")
try_except(12,0)
```

Operator Overloading**                                                                                                                                         | Using operators like `+`, `-`, `*`, etc. with **user-defined classes** by overriding special methods (`__add__`, `__str__`, etc.).          | `python class Vector:     def __init__(self, x, y):         self.x = x         self.y = y      def __add__(self, other):         return Vector(self.x + other.x, self.y + other.y)  v1 = Vector(1, 2) v2 = Vector(3, 4) v3 = v1 + v2 ` | Enables natural operations on objects (e.g., vectors, points, money, matrices).                           |
| **5. Method Overriding with `super()`** (Subtype Polymorphism)                                                                                                      | Child class extends parent behavior using **super()**, still demonstrating runtime polymorphism.                                            | `python class Shape:     def area(self):         return 0  class Rectangle(Shape):     def __init__(self, w, h):         self.w, self.h = w, h      def area(self):         return super().area() + (self.w * self.h) `                | Used when extending base class logic (e.g., UI components, API controllers, frameworks).                  |


```python
class Animal:     
    def speak(self):
        print("Animal sound")  
class Dog(Animal):     
    def speak(self):         
        print("Bark")  
a = Animal() 
a.speak()  # Animal sound 
d = Dog() 
d.speak() #Bark
```
